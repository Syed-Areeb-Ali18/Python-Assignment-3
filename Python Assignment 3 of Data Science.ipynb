```python
# Q:7 Write a program which takes 5 inputs from user for different subjectâ€™s marks, total itand generate mark sheet using grades ?
```

```python
Eng=input("Enter marks of English :");
Isl=input("Enter marks of Islamiat :");
Math=input("Enter marks of Mathematics :");
Chem=input("Enter marks of Chemistry :");
Phy=input("Enter marks of physics :");
total_marks = 500;
sum = int(Eng)+int(Isl)+int(Math)+int(Chem)+int(Phy);
print("Sum of obtained marks :",sum);
Quotient = sum / total_marks;

percentage = Quotient * 100;
print("Percentage : ",percentage);
if percentage >= 80:
    print("Grade : A-1 Grade.");
elif percentage >=70:
    print("Grade : A Grade.");
elif percentage >=60:
    print("Grade : B Grade.");
elif percentage >=50:
    print("Grade : C Grade.");
elif percentage >= 33:
    print("Grade : D Grade.");
elif percentage < 33:
    print("Grade : Fail");
```

```python
# Q: 10 Write a Python program to sum all the numeric items in a list?
```

```python
def sum_list(items):
    sum_numbers = 0
    for x in items:
        sum_numbers += x
    return sum_numbers
print(sum_list([1,2,8,3,4]))
```
